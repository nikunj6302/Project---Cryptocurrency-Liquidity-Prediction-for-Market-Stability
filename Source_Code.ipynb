```python
import pandas as pd

# Load multiple CSV files and merge them
df1 = pd.read_csv("coin_gecko_2022-03-16.csv")
df2 = pd.read_csv("coin_gecko_2022-03-17.csv")

#merge the dataframes bases on coin name with all the value assiciate with one coin in one raw
df = pd.concat([df1, df2], ignore_index=True)

# The symbol_x and symbol_y and coin name are almost same value so we will drop one of them

df = df.drop('symbol', axis=1)

df.to_csv('Merge_file.csv', index=False)

df.head()
```

Here we have given data of price, change in price and volume for different different crypto coins of the date 16th and 17th March 2022.

### Column Discription  

- Coin: this column is basicaly give the name of the crypto currency. it is a catagorical column with unique values. ALso this column is not help to predict the liquidity.

- Sumbol: This is same as 'Coin' column it has unique cat value which will not help us to predict liquidity

- Price: This is closing price price of the coin at the end of the day

- 1h, 24h, 7d : This are percentage change in the price of the cryptocurency in past respective time.

- 24h_volume: This is the exchange volume in the last 24 hours for respective crypto. This column is highly impotact to predict liquidity.

- mrt_cap: this is indicate the marckate capital of the crypto. This is also an importat feture.

- date: is the date of the treding.

In this data there are some null values but we can not do missing value treatment by using mean or median because every raw has unique value with different crypto and also we can not drop the null values because it will loss the data.
